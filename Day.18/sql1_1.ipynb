## 4. INSERT

```python
import pymysql
```

```python
products = [
    {"name": "키보드", "price": 50000},
    {"name": "마우스", "price": 25000},
    {"name": "모니터", "price": 300000},
    {"name": "노트북", "price": 1200000},
    {"name": "USB", "price": 15000},
    {"name": "웹캠", "price": 80000},
    {"name": "헤드셋", "price": 90000},
    {"name": "스피커", "price": 70000},
    {"name": "외장하드", "price": 140000},
    {"name": "프린터", "price": 220000}
]
```

```python
sql = """
insert into product(name, price)
values(%s, %s)
"""
```

```python
for i, p in enumerate(products):
    cursor.execute(sql, (p["name"], p["price"]))
    print(f"{i+1}번째 insert 실행")

conn.commit() # 테이블의 데이터 수정되는 INSERT/UPDATE/DELETE는 실행 후 반드시 commit되어야만 최종 반영됨
```

```python
sql = """
select * from product
"""

cursor.execute(sql)

rows = cursor.fetchall()
print(type(rows)) # ((), (), () ...)
print(rows)

for r in rows:
    print(r)
```

6. UPDATE

```python
# id가 1인 모니터의 price null -> 150000 으로 수정
update_price_val = 150000
update_id = 1

sql = """
update product
set price = %s 
where id = %s 
"""

cursor.execute(sql, (update_price_val, update_id))

conn.commit()

print("수정완료")
```

```python
try:

    sql = """
    delete from product
    where price is null
    """

    # db null 값을 연산할 때는 db에서 제공하는 null연산자만을 사용해야 한다.
    # is null, is not null ... 사용 / =, != 사용안됨

    cursor.execute(sql)

# select 실행 후, cursor에 반환된 결과셋을 가진다
# insert, update, delete 실행 후, 반환된 실행 행의 수 가진다

    print(f"{cursor.rowcount}행 삭제 성공")
    conn.commit()

except:
    print("db 예외발생")
finally:
    cursor.close()
    conn.close()
```
